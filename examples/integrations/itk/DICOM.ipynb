```python
from itkwidgets import view

import pooch
import itk
from pathlib import Path
```

```python
dicom_files = pooch.retrieve('https://data.kitware.com/api/v1/item/6347159711dab81428208e24/download',
                             known_hash='7ae09068c9b807b2276feb3427cebf82865ae39efc954f0222a4eec1acd6554d',
                             processor=pooch.Unzip(),
                             progressbar=True)
```

```python
# ITK will sort the DICOM series into a volume properly using spatial metadata
# Pass the directory containing the DICOM series to itk.imread
image = itk.imread(Path(dicom_files[0]).parent)
```
